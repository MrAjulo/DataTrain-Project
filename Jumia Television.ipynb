```python
import requests
```

```python
from bs4 import BeautifulSoup 
```

```python
import numpy as np
import pandas as pd
import csv
```

```python
link = 'https://www.jumia.com.ng/catalog/?q=television+'
```

```python
def scrap_page(link):
    r = requests.get(link)
    c = r.content 
    soup = BeautifulSoup(c, "html.parser")
    tv = soup.find_all("div", {"class":"sku -gallery"})
    sample = tv[0]
    
    container = []
    for sample in tv:
        con = []
        brand_name = sample.a.h2.find("span", {"class":"name"}).get_text()
        price =  sample.a.find_all("span",{"class":"price"})[0].get_text()
        con.append(brand_name)
        con.append(price)
        container.append(con)
    return container 

       
    
    
```

```python
from pprint import pprint
list_of_scrapped_pages = []
result_list = []

for i in range(1,26):
    updated_url = "https://www.jumia.com.ng/catalog/?q=television+&page=" + str(i)
    result = scrap_page(updated_url)
    pprint(result)
    answers = list_of_scrapped_pages.extend(result)
    
pprint(list_of_scrapped_pages)
```

```python
df = pd.DataFrame(list_of_scrapped_pages, columns= ['Brand_Name', 'Price'])
df
```
